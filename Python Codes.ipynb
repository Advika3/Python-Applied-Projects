```python
def run_distance_converter():
    total_meters = int(input("Enter Total Distance in Meters (e.g., 5432): "))
    if total_meters <= 0:
        print("Distance must be a non-negative number.")

    # --- PART 1: Meters to Kilometers and Remaining Meters ---
    # 1 kilometer = 1000 meters.
    # We use integer division (//) to get the whole number of kilometers.
    kilometers = total_meters // 1000
    # We use the modulus operator (%) to get the remainder in meters.
    remaining_meters_km = total_meters % 1000

    print("Kilometer Conversion:")
    print(f"Total distance: {total_meters} meters")
    print(f"Is equal to:    {kilometers} km and {remaining_meters_km} meters")

    # --- PART 2: 400m Track Rounds ---
    # We use the same operators to find how many full 400m rounds fit into the distance.
    TRACK_LENGTH = 400
    track_rounds = total_meters // TRACK_LENGTH
    remaining_meters_track = total_meters % TRACK_LENGTH

    print("400m Track Rounds:")
    print(f"Track length:   {TRACK_LENGTH} meters")
    print(f"Total rounds:   {track_rounds} full rounds")
    print(f"Remaining distance: {remaining_meters_track} meters")


# Run the function
if __name__ == "__main__":
    run_distance_converter()
```

```python
def check_prime():
    """
    Checks if a given integer n is a prime number.
    Uses the while loop optimization (checks divisors up to sqrt(n)).
    """
    print("--- Prime Number Checker ---")
    n = int(input("Enter a whole number (n): "))
    if n <= 0:
        print("Please enter a non-negative whole number.")


    if n <= 1:
        # Numbers less than or equal to 1 are not prime
        is_prime = False
    else:
        is_prime = True
        i = 2
        # Check for factors from 2 up to the square root of n
        while i * i <= n:
            if n % i == 0:
                is_prime = False
                break
            i += 1

    print("-" * 40)
    print(f"Result for {n}:")
    print(f"{n} is prime" if is_prime else f"{n} is not prime")
    print("-" * 40)

if __name__ == "__main__":
    check_prime()
```

```python
## Q9. Electricity Bill

def compute_electricity_bill(units):
    total_charge = 0
    surcharge_rate = 0.15  # 15% surcharge

    if units <= 100:
        rate = 5.00
        total_charge = units * rate
    elif units <= 300:
        # 100 units @ 5.00 + remaining units @ 7.50
        charge_slab1 = 100 * 5.00
        charge_slab2 = (units - 100) * 7.50
        total_charge = charge_slab1 + charge_slab2
    else: # units > 300
        # 100 units @ 5.00 + 200 units @ 7.50 + remaining units @ 10.00
        charge_slab1 = 100 * 5.00  # 500.00
        charge_slab2 = 200 * 7.50  # 1500.00
        charge_slab3 = (units - 300) * 10.00
        total_charge = charge_slab1 + charge_slab2 + charge_slab3

    # Apply Surcharge condition
    surcharge = 0
    if total_charge > 2000:
        surcharge = total_charge * surcharge_rate
        final_bill = total_charge + surcharge
    else:
        final_bill = total_charge

    print("\n--- Electricity Bill Summary ---")
    print(f"Units Consumed: {units}")
    print(f"Charge (before surcharge): ₹ {total_charge:,.2f}")
    print(f"Surcharge (15% if > ₹2000): ₹ {surcharge:,.2f}")
    print(f"**Total Bill Amount:** ₹ {final_bill:,.2f}")


try:
    units_consumed = int(input("Enter total units consumed: "))
    if units_consumed < 0:
        print("Units cannot be negative.")
    else:
        compute_electricity_bill(units_consumed)
except ValueError:
    print("Invalid input. Please enter a whole number for units.")
```

```python
## Q12. Parking Charge Calculator (NO IMPORTS)

RATE_FIRST_HOUR = 10.00 # Charge for the first hour
RATE_AFTER_FIRST = 5.00 # Charge for every subsequent hour

try:
    hours_parked = float(input("Enter the total hours parked (e.g., 2.5): "))
    
    if hours_parked <= 0:
        print("Time must be greater than zero.")
    else:
        # 1. Calculate Billed Hours using Basic Python Logic (The trick!)
        
        # Start with the integer part of the hours
        integer_hours = int(hours_parked)
        
        # Check if there is any fractional part left
        if hours_parked > integer_hours:
            # If time is 2.1, integer_hours is 2. We add 1 to round up to 3.
            billed_hours = integer_hours + 1
        else:
            # If time is 2.0, integer_hours is 2. No rounding up needed.
            billed_hours = integer_hours
        
        
        charge_first_hour = 0
        charge_remaining = 0
        
        if billed_hours <= 1:
            # Case 1: 1 hour or less billed
            charge_first_hour = RATE_FIRST_HOUR
            total_charge = charge_first_hour
            remaining_billed_hours = 0
        else:
            # Case 2: More than 1 hour billed
            
            charge_first_hour = RATE_FIRST_HOUR
            remaining_billed_hours = billed_hours - 1
            charge_remaining = remaining_billed_hours * RATE_AFTER_FIRST
            
            total_charge = charge_first_hour + charge_remaining

        # --- Output ---
        print("\n" + "="*40)
        print(f"Time Parked: {hours_parked:.2f} hrs")
        print(f"Billed Hours: {billed_hours} hrs (Rounded Up)")
        print("-" * 40)
        
        print(f"1st Hour Charge:  ₹ {charge_first_hour:,.2f}")
        if remaining_billed_hours > 0:
             print(f"Remaining {remaining_billed_hours} hrs: ₹ {charge_remaining:,.2f}")
        
        print("-" * 40)
        print(f"**Total Charge:** ₹ {total_charge:,.2f}")
        print("="*40)

except ValueError:
    print("Error: Input must be a valid number.")
```

```python
## Q15. Digit Frequency Counter

try:
    number = int(input("Enter an integer: "))
except ValueError:
    print("Invalid input.")
    exit()

temp_num = abs(number) # Use the absolute value to handle negatives
frequencies = {}       # {digit: count}

if temp_num == 0:
    frequencies[0] = 1 # Handle the input 0

while temp_num > 0:
    # 1. Get the last digit
    digit = temp_num % 10  
    
    # 2. Add 1 to the count for that digit (or set it to 1 if first time)
    frequencies[digit] = frequencies.get(digit, 0) + 1
    
    # 3. Remove the last digit
    temp_num = temp_num // 10 

print(f"\nFrequencies in {number}:")
# Sort the output by digit for neatness
for digit in sorted(frequencies.keys()):
    print(f"Digit {digit}: {frequencies[digit]} times")
```

```python
## Q16. Fibonacci Series

try:
    n_terms = int(input("Enter the number of Fibonacci terms to generate: "))
except ValueError:
    print("Invalid input.")
    exit()

a, b = 0, 1       # Start terms
sum_even = 0
sum_odd = 0
count = 0

print("Fibonacci Series:")

while count < n_terms:
    print(a, end=" ")
    
    # Check if 'a' (the current term) is even or odd
    if a % 2 == 0:
        sum_even += a
    else:
        sum_odd += a

    # Update for the next iteration (a becomes the old b, b becomes the new sum)
    a, b = b, a + b
    count += 1

print(f"\n\nSum of Even terms: {sum_even}")
print(f"Sum of Odd terms: {sum_odd}")
```

```python
## Q18. Word Frequency Counter

import string             # Tool for getting punctuation list
from collections import Counter # FAST tool for counting things

FILE_NAME = "story.txt"

# --- STEP 0: Make sure the file exists (Create a quick dummy file) ---
# You'll need this part to run the code the first time
DUMMY_STORY = """
The fox runs fast. The dog runs slow. The dog and the fox are running.
"""
with open(FILE_NAME, 'w') as f:
    f.write(DUMMY_STORY)
print(f"File '{FILE_NAME}' created.")


def compute_word_frequency():
    try:
        # 1. READ
        with open(FILE_NAME, 'r') as file:
            text = file.read()

        # 2. CLEANING (The important stuff)
        text = text.lower() # Make everything lowercase

        # Remove punctuation (.,!?)
        translator = str.maketrans('', '', string.punctuation)
        clean_text = text.translate(translator)

        # Split into words (splits by spaces and removes empty strings)
        words = clean_text.split()
        
        # 3. COUNT
        word_counts = Counter(words)

        # 4. TOP 5
        top_5 = word_counts.most_common(5)

        print("-" * 50)
        print("--- Top 5 Word Frequencies ---")
        
        for rank, (word, count) in enumerate(top_5, 1):
            print(f"{rank}. '{word}': {count} times")

    except FileNotFoundError:
        print(f"Error: Couldn't find the file '{FILE_NAME}'.")

compute_word_frequency()
```

```python
## Q17. Student Records File

FILE_NAME = "student_records.txt"
MIN_PASS_MARK = 75

def write_records():
    """Writes N student records (Name, Roll, Marks) to the file."""
    try:
        n = int(input("Enter how many students to record: "))
        if n <= 0:
            print("Gotta have at least one student.")
            return

        # Open file in 'w' (write) mode. This overwrites the file every time.
        with open(FILE_NAME, 'w') as file:
            print("\n--- ENTERING RECORDS ---")
            for i in range(n):
                print(f"Record {i+1}:")
                name = input("  Name: ")
                roll = input("  Roll No: ")
                # Error handling to make sure marks is a number
                while True:
                    try:
                        marks = int(input("  Marks (0-100): "))
                        break
                    except ValueError:
                        print("    Marks gotta be a number, man.")
                
                # Write to file: Format is Name,Roll,Marks\n
                file.write(f"{name},{roll},{marks}\n")
        
        print(f"\nSaved {n} records to {FILE_NAME}.")

    except ValueError:
        print("Invalid input for student count.")

def read_and_filter_records():
    """Reads the file and prints students with marks >= 75."""
    print(f"\n--- ELIGIBLE STUDENTS (Marks >= {MIN_PASS_MARK}) ---")
    found_eligible = False
    
    try:
        # Open file in 'r' (read) mode
        with open(FILE_NAME, 'r') as file:
            
            # Loop through the file line by line
            for line in file:
                data = line.strip() # Removes the '\n' at the end
                if not data: continue # Skip if the line is empty

                try:
                    # Split the line by the comma: data -> [Name, Roll, Marks]
                    name, roll, marks_str = data.split(',')
                    marks = int(marks_str) # Convert marks back to integer
                    
                    # Check the condition
                    if marks >= MIN_PASS_MARK:
                        print(f"Name: {name:<10} | Roll No: {roll:<5} | Marks: {marks}")
                        found_eligible = True
                
                except Exception:
                     # This catches bad formatting or non-integer marks
                     print(f"Warning: Skipping bad record: {data}")
        
        if not found_eligible:
            print(f"No students found with marks above {MIN_PASS_MARK}.")
            
    except FileNotFoundError:
        print(f"Error: The file '{FILE_NAME}' isn't there. Run the writing function first.")

# Run both parts of the assignment
write_records()
read_and_filter_records()
```

```python
## Q19. Array Statistics

import numpy as np

# Create a sample NumPy array
data_array = np.array([85, 12, 98, 55, 78, 98, 63, 15, 70, 42])
print(f"Original Array: {data_array}")
print("-" * 30)

# 1. Mean (Average)
print(f"Mean (Average):     {np.mean(data_array):.2f}")

# 2. Median (Middle value)
print(f"Median:             {np.median(data_array):.2f}")

# 3. Standard Deviation (Spread)
print(f"Std Dev:            {np.std(data_array):.2f}")

# 4. Maximum Value
print(f"Maximum Value:      {np.max(data_array)}")

# 5. Minimum Value
print(f"Minimum Value:      {np.min(data_array)}")

# 6. Index of Maximum Value (first occurrence)
print(f"Index of Max Value: {np.argmax(data_array)}")
```

```python
## Q20. Matrix Operations

import numpy as np

# Define two 2x3 matrices for addition and element-wise multiplication
matrix_A = np.array([[1, 2, 3], 
                     [4, 5, 6]])

matrix_B = np.array([[7, 8, 9], 
                     [10, 11, 12]])

# Define a third matrix (3x2) for the dot product
# (A is 2x3, C is 3x2, result is 2x2)
matrix_C = np.array([[1, 0], 
                     [0, 1], 
                     [1, 1]])

print("Matrix A:\n", matrix_A)
print("Matrix C:\n", matrix_C)
print("-" * 30)

# 1. Matrix Addition (Element-wise)
# Requires A and B to have the same dimensions (A + B)
addition_result = matrix_A + matrix_B
print("1. Matrix Addition (A + B):\n", addition_result)

# 2. Dot Product (True Matrix Multiplication)
# Uses np.dot() or the @ operator (A @ C)
dot_product_result = np.dot(matrix_A, matrix_C)
print("\n2. Dot Product (A * C):\n", dot_product_result)

# 3. Element-wise Multiplication (Hadamard Product)
# Uses the standard * operator (A * B)
element_wise_result = matrix_A * matrix_B
print("\n3. Element-wise Multiplication (A * B):\n", element_wise_result)
```

```python
## Q21. NumPy Advanced Indexing

import numpy as np

# 1. Generate 15 random scores (0 to 100)
# np.random.seed() ensures you get the same 'random' scores every time
np.random.seed(42) 
scores = np.random.randint(0, 101, 15)
print(f"Original 15 Scores:\n{scores}")
print("-" * 50)

# 2. Boolean Filter: Select scores >= 50
# Create the boolean mask
mask_pass = scores >= 50
print(f"Boolean Mask (scores >= 50):\n{mask_pass}")

# Use the mask to select and print the passing scores
passing_scores = scores[mask_pass]
print(f"\nPassing Scores (>= 50):\n{passing_scores}")
print("-" * 50)

# 3. Replace scores < 40 with 40
# Create a new mask for scores that are failing
mask_fail = scores < 40

# Use the mask to assign the value 40 to only those elements
scores[mask_fail] = 40

print("Final Modified Scores:")
print(f"Original scores < 40 are now 40:\n{scores}")
```

```python
## Q22. Pandas Series - Sales Analysis

import pandas as pd

# Data: June to November sales
months = ['Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov']
current_sales = [25000, 28000, 31000, 35000, 33000, 40000]
previous_sales = [23000, 27000, 30000, 36000, 32000, 38000]

# Create two Series objects
sales_current = pd.Series(current_sales, index=months, name='Current Year')
sales_previous = pd.Series(previous_sales, index=months, name='Previous Year')

# Combine into a DataFrame to easily calculate the difference
sales_df = pd.DataFrame({'Current': sales_current, 'Previous': sales_previous})

# Calculate the difference and growth
sales_df['Difference'] = sales_df['Current'] - sales_df['Previous']
sales_df['Growth (%)'] = ((sales_df['Difference'] / sales_df['Previous']) * 100).round(2)

print("--- Sales Comparison ---")
print(sales_df)
print("\nMonths where Current Sales were HIGHER:")
# Use boolean indexing on the DataFrame
print(sales_df[sales_df['Difference'] > 0].index.tolist())
```

```python
## Q23. DataFrame - Student Result Table

import pandas as pd

data = {
    'Roll No': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Phy': [65, 88, 35, 75, 92],
    'Chem': [70, 95, 40, 80, 88],
    'Math': [60, 90, 25, 65, 95]
}
df = pd.DataFrame(data)

# 1. Add 'Total' Column
df['Total'] = df['Phy'] + df['Chem'] + df['Math']

# 2. Add 'Result' Column (Pass if all subjects >= 40)
# Use a lambda function with .apply() to check all columns in one row
df['Result'] = df.apply(
    lambda row: 'Pass' if all(row[s] >= 40 for s in ['Phy', 'Chem', 'Math']) else 'Fail', 
    axis=1 # axis=1 applies the function row-wise
)

print("--- Student Result Table ---")
print(df)

# 3. Filter Pass Students
passed_df = df[df['Result'] == 'Pass']
print("\n--- Passed Students ---")
print(passed_df)

# 4. Print Topper (using idxmax() to get the index of the max total)
topper = df.loc[df['Total'].idxmax()]
print("\n--- Class Topper ---")
print(f"Name: {topper['Name']}, Total: {topper['Total']}")
```
